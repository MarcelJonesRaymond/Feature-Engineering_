```python
y=data[['diagnosis']]
x=data.iloc[:,1:]
```

```python
from scipy.stats import zscore
scaled_x=x.apply(zscore)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
Y=lb.fit_transform(y)
```

```python
Y=pd.DataFrame(Y,columns=['diagnosis'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold(n_splits=10)
for train,test in cv.split(scaled_x,Y):
    x_train,x_test=x.loc[train],x.loc[test]
    y_train,y_test=y.loc[train],y.loc[test]
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
Imp=pd.DataFrame(dt.feature_importances_,index=x_train.columns,columns=['Importance']).sort_values(by='Importance',ascending=False)
Imp_val=Imp.head(14).index
```

```python
Imp_val
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train[Imp_val],y_train)
yhat=dt.predict(x_test[Imp_val])
```

```python
print(metrics.classification_report(y_test,yhat))
```

```python
print(metrics.confusion_matrix(y_test,yhat))
```

```python
from sklearn.model_selection import cross_val_score
result=cross_val_score(dt,x_train,y_train,cv=StratifiedKFold(n_splits=10))
print(' Therefore the Average value that can be expected from this model is that {} with standdard deviation of {} '.format(round(result.mean()*100),4),round(result.std()*100,4))
```
